#### 3.1.3 Which quadrant has higher average median household income across different households types?

```python
engine = sq.create_engine('mysql+mysqlconnector://l01-10:BhZC3ULTKA9MA@datasciencedb.ucalgary.ca/l01-10')

engine.execute('use `l01-10`;')

# 1)
engine.execute('''
CREATE TEMPORARY TABLE median_income
SELECT 
    c.comm_name,
    c.ct,
    c.quadrant,
    c.comm_code,
    c.class,
    m.household_size,
    m.household_type,
    m.household_income_statistics,
    `value`
FROM
    calgary_communities c
        JOIN
    median_household_income m ON c.ct = m.ct;
''')

# 2)
engine.execute('''
CREATE TEMPORARY TABLE dup_ct_labels 
SELECT 
    m.comm_code,
    IFNULL(mt.is_duplicate, 'duplicate') AS is_duplicate
FROM
    median_income m
        LEFT JOIN
    (SELECT 
        ct, comm_code, 'ct_unique' AS is_duplicate
    FROM
        median_income
    GROUP BY ct) AS mt ON m.comm_code = mt.comm_code
GROUP BY m.comm_code;
''')

# 3)
engine.execute('''
CREATE TEMPORARY TABLE mhi_totals
SELECT 
        *
    FROM
        median_income
    WHERE
        household_type = 'Total Household type including census family structure'
            AND household_size = 'Total - Households by household size'
            AND household_income_statistics = 'Median household total income'
''')

# 4)
engine.execute('''
CREATE TEMPORARY TABLE mhi_totals_summary
SELECT 
    mt.quadrant,
    ROUND(AVG(CASE
                WHEN d.is_duplicate = 'ct_unique' THEN mt.`value`
                ELSE NULL
            END)) AS avg_mhi_totals
FROM
    mhi_totals mt
        JOIN
    dup_ct_labels d ON mt.comm_code = d.comm_code
GROUP BY mt.quadrant;
''')

# 5)
engine.execute('''
CREATE TEMPORARY TABLE mhi_household_profiles
SELECT 
    quadrant,
        ROUND(AVG(CASE
                WHEN
                    (mi.household_type = 'Census-family households'
                        AND d.is_duplicate = 'ct_unique')
                THEN
                    mi.`value`
                ELSE NULL
            END)) AS avg_mhi_censusfamily,
    ROUND(AVG(CASE
                WHEN
                    (mi.household_type = 'Without Children'
                        AND d.is_duplicate = 'ct_unique')
                THEN
                    mi.`value`
                ELSE NULL
            END)) AS avg_mhi_nochildren,
    ROUND(AVG(CASE
                WHEN
                    (mi.household_type = 'With Children'
                        AND d.is_duplicate = 'ct_unique')
                THEN
                    mi.`value`
                ELSE NULL
            END)) AS avg_mhi_withchildren,
    ROUND(AVG(CASE
                WHEN
                    (mi.household_type = 'One one-parent census family'
                        AND d.is_duplicate = 'ct_unique')
                THEN
                    mi.`value`
                ELSE NULL
            END)) AS avg_mhi_oneparent,
    ROUND(AVG(CASE
                WHEN
                    (mi.household_type = 'Other census family households'
                        AND d.is_duplicate = 'ct_unique')
                THEN
                    mi.`value`
                ELSE NULL
            END)) AS avg_mhi_other,
    ROUND(AVG(CASE
                WHEN
                    (mi.household_type = 'Non-census family households'
                        AND d.is_duplicate = 'ct_unique')
                THEN
                    mi.`value`
                ELSE NULL
            END)) AS avg_mhi_noncensus
FROM
    median_income mi
        JOIN
    dup_ct_labels d ON mi.comm_code = d.comm_code
WHERE
    household_size = 'Total - Households by household size'
        AND household_income_statistics = 'Median household total income'
GROUP BY quadrant; 
''')

# 6)
income_query = '''
SELECT
    h.*,
    avg_mhi_totals
FROM
    mhi_totals_summary t
        JOIN
    mhi_household_profiles h ON t.quadrant = h.quadrant
'''

income = pd.read_sql_query(income_query, engine)
display(income)

engine.dispose()
```

Results:

The table shows that the average median household income in the Southwest is greater than the average in the median household income in the Northeast across all household types. Both quadrants show higher average median household incomes among households with children, while the lowest averages are observed among non-census family households. This pattern underscores the importance of family composition in shaping economic conditions within these regions.

### 3.2.3 What is the relationshiop between ownership and total number of crimes? 

One of the primary objectives of our analysis is to examine the homeownership rates in the Northeast and Southwest regions, and to investigate the potential impact of these rates on crime statistics. Homeownership is regarded as a significant economic indicator, and understanding its influence on crime rates can provide valuable insights for policymakers and community planners. To facilitate this analysis, we have joined the demographic dataset with the crime statistics dataset obtained from the Calgary Police Services on SQL. To investigate the correlation between the total number of homeowners and the incidence of crimes, we developed a series of queries:

Steps:

1. The first query extracts data regarding various communities, incorporating both demographic and crime statistics, and subsequently filters out any communities with zero residents or homeowners. This query merges crime and census data on a community-specific basis, calculates the total crimes committed in each community for the year 2020, and generates detailed reports for each community. 
2. The second query refines this data to focus specifically on the most relevant columns—namely, the number of homeowners and total crimes—excluding other data elements that are unnecessary for our analysis. 
3. In the third part, we have constructed a query to manually calculate the Pearson Correlation Coefficient. This step is crucial as it allows us to determine the statistical correlation between homeownership rates and crime rates directly within SQL, providing a clear and precise measure of their interdependence.

```python
engine = create_engine('mysql+mysqlconnector://l01-10:BhZC3ULTKA9MA@datasciencedb.ucalgary.ca/l01-10')


query1 = """
SELECT * FROM (
    SELECT 
        communityname, 
        census_data.section,
        dwellings,
        preschool_children,
        homeowners,
        residents,
        male_residents,
        female_residents,
        SUM(`2020`) AS total_crimes
    FROM 
        crime_analysis 
        JOIN census_data ON crime_analysis.communityname = census_data.community_name
    GROUP BY 
        communityname
    ORDER BY 
        communityname
) AS subquery
WHERE 
    residents > 0 AND homeowners > 0;
"""

query2 = """
SELECT 
    communityname, 
    homeowners,
    SUM(`2020`) AS total_crimes
FROM 
    crime_analysis 
    JOIN census_data ON crime_analysis.communityname = census_data.community_name
GROUP BY 
    communityname
HAVING 
    homeowners > 0;
"""

query3 = """
SELECT 
    SUM((homeowners - avg_homeowners) * (total_crimes - avg_total_crimes)) / 
    (SQRT(SUM(POW(homeowners - avg_homeowners, 2))) * SQRT(SUM(POW(total_crimes - avg_total_crimes, 2)))) AS correlation_coefficient
FROM (
    SELECT 
        homeowners,
        total_crimes,
        AVG(homeowners) OVER () AS avg_homeowners,
        AVG(total_crimes) OVER () AS avg_total_crimes
    FROM (
        SELECT 
            communityname, 
            homeowners,
            SUM(`2020`) AS total_crimes
        FROM 
            crime_analysis 
            JOIN census_data ON crime_analysis.communityname = census_data.community_name
        GROUP BY 
            communityname
        HAVING 
            homeowners > 0
    ) AS raw_data
) AS derived_data;
"""

def fetch_query_results(query, engine):
    return pd.read_sql_query(query, engine)

df1 = fetch_query_results(query1, engine)
df2 = fetch_query_results(query2, engine)
df3 = fetch_query_results(query3, engine)

# Now, you can print the DataFrames or work with them further:
print("Results from Query 1:")
display(df1)
print("\nResults from Query 2:")
display(df2)
print("\nResults from Query 3:")
display(df3)

engine.dispose()
```

Results:

Upon completion of the correlation analysis, the calculated correlation coefficient was 0.32, signifying a moderate positive relationship between the number of homeowners and total crimes within the dataset. Although the correlation is not exceptionally robust, it implies that as homeownership in a community increases, there is a concurrent tendency for the total number of crimes to rise. It is important to emphasize, however, that while a positive relationship exists, this correlation should not be interpreted as causation. An increase in homeownership does not necessarily cause an increase in crime rates directly. Other underlying factors, such as population density, could be influencing both high rates of homeownership and higher crime rates.

### 3.2.4 How does lower median household income influence the incidence of different crimes?

Perception suggests a correlation between lower income and higher crime numbers. Notably, Non-census families, which is one person living alone or a group of two or more persons who live together but do not constitute a census family (Statistics Canada), had the lowest average income. Therefore, we will investigate the distribution of four crime types: Assault (Non-domestic), Violence 'Other' (Non-domestic), Break & Enter - Dwelling, and Theft OF Vehicle'. This analysis will consider the proportion of Non-census families to total households per census tract. A higher proportion indicates more households with lower income, 
potentially contributing to a rise in criminal activity.

Steps:

1. Inner join "calgary_communities" with "medium_household_income" using census tract (ct) and create temporary table "median_income".

2. Filter "median_income" by total household sizes, non-census family in household type, change the household statistic from income to number of households, and group by census tract (ct). Create temporary table "t_noncensusf_households".

3. Filter "median_income" by totals in size and type of households, and create temporary table "mhi_totals", change the household statistic from income to number of households, inner join table "t_noncensusf_households" using census tract (ct), group by census tract (ct) and quadrant, compute ratio of non-census households to total houses and rank by ratio in descendant order.

4. Inner join "median_income" with "crime_analysis" using community name, and "noncensusf_ratio_ranking" using census tract (ct). Filter by non-census family in household type, total sizes, and change change the household statistic from income to number of households. Filter crimes by four types. Exclude census tract 0007.05 since does not include information for three of the selected crime types. Group by census tract and crime category, order by quadrant and ranking. Display results.

```python
engine = sq.create_engine('mysql+mysqlconnector://l01-10:BhZC3ULTKA9MA@datasciencedb.ucalgary.ca/l01-10')
engine.execute('use `l01-10`;')

# 1)
engine.execute('''
CREATE TEMPORARY TABLE median_income
SELECT 
    c.comm_name,
    c.ct,
    c.quadrant,
    c.comm_code,
    c.class,
    m.household_size,
    m.household_type,
    m.household_income_statistics,
    `value`
FROM
    calgary_communities c
        JOIN
    median_household_income m ON c.ct = m.ct;
''')

# 2)
engine.execute('''
CREATE TEMPORARY TABLE t_noncensusf_households
SELECT 
    ct, `value` AS noncensusf_households
FROM
    median_income
WHERE
    household_income_statistics = 'Number of households'
        AND household_type = 'Non-census family households'
        AND household_size = 'Total - Households by household size'
GROUP BY ct;
''')

# 3)
engine.execute('''
CREATE TEMPORARY TABLE noncensusf_ratio_ranking
SELECT 
    mi.ct,
    mi.quadrant,
    noncensusf_households, 
    `value` AS total_households, 
    noncensusf_households / `value` AS ratio,
    RANK() OVER (PARTITION BY mi.quadrant ORDER BY ratio DESC) AS ranking
FROM
    median_income mi
        JOIN t_noncensusf_households t ON mi.ct = t.ct
WHERE
    household_income_statistics = 'Number of households'
        AND household_type = 'Total Household type including census family structure'
        AND household_size = 'Total - Households by household size'
GROUP BY mi.ct
ORDER BY quadrant, ranking;
''')

# 4)
g_question_query = '''
SELECT 
    mi.ct,
    mi.quadrant,
    `value` AS mhi,
    ca.Category crime_category,
    SUM(ca.`2020`) AS crime_count,
    noncensusf_households, 
    total_households, 
    ratio,
    ranking
FROM
    median_income mi
        JOIN
    crime_analysis ca ON mi.comm_name = ca.CommunityName
        JOIN
    noncensusf_ratio_ranking r on mi.ct = r.ct
WHERE
    household_income_statistics = 'Median household total income'
        AND household_type = 'Non-census family households'
        AND household_size = 'Total - Households by household size'
        AND (Category = 'Assault (Non-domestic)'
        OR Category = "Violence  'Other' (Non-domestic)"
        OR Category = 'Theft OF Vehicle'
        OR Category = 'Break & Enter - Dwelling')
        AND mi.ct != '0007.05'
GROUP BY mi.ct , crime_category
ORDER BY mi.quadrant, ranking;

'''
g_question = pd.read_sql_query(g_question_query, engine)
display(g_question)

engine.dispose()
```
